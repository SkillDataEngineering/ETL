## 5. During a search, how often is the driver frisked?

```python
# Bucket the unique values for a count!
ri.search_type.value_counts()
```

```python
# Naive approach

# We create a new column, expecting it to be True or False depending on whether a Frisk was performed:

frisk_conducted = ri.search_type == 'Protective Frisk'
ri['frisk'] = frisk_conducted
```

```python
# The above won't work well. It turns NaN values into False:s and also fails to capture cells where 'Protective Frisk' was part of a longer string.

# string columns have a super-useful .str attribute loaded with useful functionality though!
# The following preserves missing values while searching for the string or substring!
ri['frisk'] = ri.search_type.str.contains('Protective Frisk')
```

```python
# Better now
ri['frisk'].value_counts(dropna=False)
```

```python
# rate = instances / total
# Equivalents

frisk_rate = ((ri.frisk == True).sum()) / ((ri.frisk == True).sum() + (ri.frisk == False).sum())
frisk_rate2 = ((ri.frisk == True).sum()) / (ri.frisk.notna().sum())
print(frisk_rate)
print(frisk_rate2)
```

## Injection: Datetime type

```python
# Make a datetime type column "stop_datetime", from the existing str type columns "stop_date" & "stop_time"
# cat not as in the animal, but short for "conCATenate"
combined = ri.stop_date.str.cat(ri.stop_time, sep=' ')
ri['stop_datetime'] = pd.to_datetime(combined)
```

```python
# We can now access useful attributes such dt.year or dt.hour within a column of datetime type.
ri.stop_datetime.dt.year.value_counts()
```

## 6. Find the bad data in the stop_duration column and fix it

- What is the "bad data"?

```python
ri.stop_duration.value_counts(dropna=False)
```

- 1 and 2 are the values we consider bad.

```python
# Plan: Replace the cells that have values 1 and 2 with NaN (missing value)
```

```python
# Stop Duration is an 'O'bject type (string) column
ri.stop_duration.dtype
```

```python
# Wishful thinking ... but it is what we want to do. Just need the correct syntax.

# ri[ri.stop_duration == 1 or ri.stop_duration == 2] = 'NaN'
```

```python
# This won't work because of the method chaining of Dataframe subselection/views/copies
# Selecting on rows returns a view, and then in that subselection we select on the stop_duration column.
# If we try to assign, we're assigning to a copy of a copy that can't reference back to the original 'ri' dataframe.
ri[ri.stop_duration == '1'].stop_duration = 'temporary_value'

# Thus the "SettingWithCopyWarning". Pandas recognizes that what you're TRYING to do is not what IT WILL do.
# It proposes that we use .loc[row_indexing, col_indexing] to find the right cell(s) instead.
```

```python
# The default NaN-values in Pandas are the special nan values from the NumPy module: np.nan
import numpy as np
```

```python
# Correct!

# | = logical or
# & = logical and
ri.loc[
    (ri.stop_duration == '1') | (ri.stop_duration == '2'),  # row indexing
    'stop_duration'                                         # column indexing
    ] = np.nan
```
