### 2. When men and women get pulled over, how often is the reason **Speeding**?

Steps to reach our goal?
- We are interested in rows for which the "violation" (or "violation_raw") value is *Speeding*
- Somehow count up, sum, or separate driver_gender [ *M* / *F* ] occurences?

```python
speedsters = ri.groupby('violation')
ri_speedsters = speedsters.get_group('Speeding')
speeding = ri_speedsters.driver_gender.value_counts()

totals = ri.driver_gender.value_counts()

male_ratio = speeding['M']/totals['M']
female_ratio = speeding['F']/totals['F']

print(male_ratio, female_ratio)
```

```python
speedings = ri[ri['violation_raw'] == 'Speeding'][:]
female = speedings['driver_gender'].value_counts()['F']
male = speedings['driver_gender'].value_counts()['M']
 
print("Females speed {} times".format(female))
print("Males speed {} times".format(male))
```

```python
ri.groupby('driver_gender').violation.value_counts(normalize=True)
```
